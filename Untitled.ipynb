```python
with open('pos', 'rb') as handle:
    dict = pickle.load(handle)
```

```python
dict
```

```python
df=pd.DataFrame.from_dict(dict,orient="index")
```

```python
df=df.melt(value_vars=df.columns,value_name="position",var_name="agent_id")
```

```python
df["x_pos"]=df.position.apply(lambda x: x[0])
df["y_pos"]=df.position.apply(lambda x: x[1])
```

```python
df=df.drop(["position"],axis=1)
df["step"] = list(range(5))*10
```

```python

df['diff_x'] = df.groupby("agent_id")[["x_pos"]].diff().fillna(0)
df["diff_y"]= df.groupby("agent_id")[["y_pos"]].diff().fillna(0)
df["move"]=np.sqrt(df["diff_x"]**2+df["diff_y"]**2)
```

```python
df.groupby("step")["move"].mean().plot().get_figure().savefig('output.png')
```
